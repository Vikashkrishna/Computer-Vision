```python
import cv2
from google.colab.patches import cv2_imshow
```

LINEAR SPATIAL FILTERING USING BUILT -IN FUNCTIONS

```python
image = cv2.imread('/content/cat.jpg')
```

```python
box_filt = cv2.boxFilter(image,-1,(5,5),borderType=cv2.BORDER_DEFAULT)
```

```python
gau_filt = cv2.boxFilter(image,-1,(5,5),borderType=cv2.BORDER_DEFAULT)
```

```python
print("Original Image")
cv2_imshow( image)
print("Image after Box Filter")
cv2_imshow( box_filt)
print("Image after Gaussian Blur")
cv2_imshow( gau_filt)
```

SHARPENING

```python
image = cv2.imread('/content/cat.jpg',0)
```

```python
import numpy as np
laplacian = cv2.Laplacian(image, cv2.CV_64F)
laplacian_uint8 = np.uint8(np.absolute(laplacian))
sharpened_image = cv2.addWeighted(image, 0.8, laplacian_uint8, -0.5, 0)
print("Original Image")
cv2_imshow(image)
print("Sharpened Image")
cv2_imshow(sharpened_image)
```

USING USER DEFINED FUNCTIONS

```python
box_blur_kernel = np.array([[1, 1, 1],
                            [1, 1, 1],
                            [1, 1, 1]]) / 9
gaussian_blur_kernel = np.array([[1, 2, 1],
                                 [2, 4, 2],
                                 [1, 2, 1]]) / 16
box_blur_kernel_result = cv2.filter2D(image, -1, box_blur_kernel)
gaussian_blur_kernel_result = cv2.filter2D(image, -1, gaussian_blur_kernel)
```

```python
print("Original Image")
cv2_imshow(image)
print("Box Blur Filtering")
cv2_imshow(box_blur_kernel_result)
print("Gaussian Blur Filtering")
cv2_imshow(gaussian_blur_kernel_result)
```

```python
laplacian_sharpen_kernel = np.array([[0, -1, 0],
                                     [-1, 5, -1],
                                     [0, -1, 0]])
laplacian_sharpen_kernel_result = cv2.filter2D(image, -1, laplacian_sharpen_kernel)
```

```python
print("Laplacian Sharpen Kernel Result (Laplacian Sharpening)")
cv2_imshow(laplacian_sharpen_kernel_result)
```
